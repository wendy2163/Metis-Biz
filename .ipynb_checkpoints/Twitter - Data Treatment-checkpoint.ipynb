```python
df_tw=pd.read_csv('C:\\Users\wangz\metis\Business\Twitter_project\Tweet.csv')
df_tw.head()
```

```python
import datetime
df_tw['date']=pd.to_datetime(df_tw['post_date'], unit='s')
tweet = df_tw.drop(['post_date'], axis=1, inplace=False)
```

```python
tweet['year'] = tweet['date'].dt.year
tweet.head()
```

```python
tweet = tweet.dropna()
```

```python
y=tweet[['year']]
y.value_counts()
```

```python
tweet2019=tweet[tweet['year'] == 2019]
tweet2019.head()
```

```python
print(f"Number of writers: {tweet2019['writer'].nunique()}")
```

```python
stats = tweet2019[['writer', 'tweet_id']].groupby('writer').agg("count").rename(columns={'tweet_id' : 'tweet_count'})
```

```python
sns.set_context("paper", font_scale=2)

plt.figure(figsize=(12, 8))

sns.histplot(data=stats, x='tweet_count', bins=50, log_scale=True)
plt.yscale('log')
plt.title("Posts count histogram")
plt.xlabel("Number of users")
plt.ylabel("Tweet count")
None
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
sentiment_nltk = SentimentIntensityAnalyzer()

# Mild cleaning: remove weblinks, $ticker_symbol, # symbol from hashtags, remove excessive spaces
tweet2019['prep_body'] = tweet2019['body'].replace(r"https?:\S+|http?:\S+|www?:\S+", '', regex=True).replace(r"[@#\$][a-zA-Z]+", '', regex=True).replace(r"\s\s+", ' ', regex=True).str.strip()

tweet2019['positive_sentiment'] = tweet2019['prep_body'].apply(lambda x: sentiment_nltk.polarity_scores(x)['pos'])
tweet2019['negative_sentiment'] = tweet2019['prep_body'].apply(lambda x: sentiment_nltk.polarity_scores(x)['neg'])
tweet2019['total_sentiment'] = tweet2019['prep_body'].apply(lambda x: sentiment_nltk.polarity_scores(x)['compound'])

tweet2019.head()
```

```python
# calculate # of duplicates 
print("Percent of duplicated tweets: {:.2f} %".format(sum(tweet2019['body'].duplicated())/len(tweet2019) * 100))
```

```python
tweet2019['flag_potential_bot']=tweet2019['body'].duplicated()
```
